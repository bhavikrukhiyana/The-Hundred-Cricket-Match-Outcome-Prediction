```python
# Randomizing each column's values independently
cric_df['Random Runs Scored'] = np.random.permutation(cric_df['Number of Runs Scored'])
cric_df['Random Balls Faced'] = np.random.permutation(cric_df['Number of Balls Faced'])

# Ensure correct datatypes
cric_df['Date'] = pd.to_datetime(cric_df['Date'])

# Function to add regression line
def add_regression_line(x, y, line_color):
    # Remove NaN and infinite values
    mask = np.isfinite(x) & np.isfinite(y)
    x, y = x[mask], y[mask]
    
    # Normalize the data
    x_norm = (x - np.mean(x)) / np.std(x)
    y_norm = (y - np.mean(y)) / np.std(y)
    
    coeffs = np.polyfit(x_norm, y_norm, 1)
    poly = np.poly1d(coeffs)
    
    # Generate regression line
    x_seq = np.linspace(min(x), max(x), num=100)
    plt.plot(x_seq, poly((x_seq - np.mean(x)) / np.std(x)) * np.std(y) + np.mean(y), color=line_color)

# Scatter plot for original data
plt.figure(figsize=(12, 8))
scatter_colors = {'male': 'blue', 'female': 'pink'}
line_colors = {'male': 'green', 'female': 'red'}

for gender in cric_df['Gender'].unique():
    subset = cric_df[cric_df['Gender'] == gender]
    runs_per_ball = subset['Number of Runs Scored'] / subset['Number of Balls Faced']
    plt.scatter(subset['Number of Runs Scored'], runs_per_ball, label=gender, alpha=0.6, color=scatter_colors[gender])
    add_regression_line(subset['Number of Runs Scored'], runs_per_ball, line_colors[gender])

plt.xlabel('Number of Runs Scored')
plt.ylabel('Runs Per Ball (Number of Runs Scored / Number of Balls Faced)')
plt.title('Original Data: Runs Scored vs Runs Per Ball by Gender with Regression Lines')
plt.legend(title='Gender')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.nonparametric.smoothers_lowess import lowess

# Calculate the ratios
cric_df['Random Runs per Ball'] = cric_df['Random Runs Scored'] / cric_df['Random Balls Faced']
cric_df['Runs Per Ball'] = cric_df['Number of Runs Scored'] / cric_df['Number of Balls Faced']

# Function to plot data with a smooth curve using LOWESS
def plot_with_curve(x, y, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, label='Data points')
    
    # Apply LOWESS smoothing
    smoothed = lowess(y, x, frac=0.3)
    
    plt.plot(smoothed[:, 0], smoothed[:, 1], color='red', label='Smooth curve')
    plt.title(title)
    plt.xlabel('Balls Faced')
    plt.ylabel('Runs per Ball')
    plt.legend()
    plt.grid(True)
    # plt.savefig(f'{title}.png')
    plt.show()


# Plot 1: Random Balls Faced vs Random Runs per Ball
plot_with_curve(cric_df['Random Balls Faced'], cric_df['Random Runs per Ball'], 'Random Balls Faced vs Random Runs per Ball')

# Plot 2: Number of Balls Faced vs Runs per Ball
plot_with_curve(cric_df['Number of Balls Faced'], cric_df['Runs Per Ball'], 'Number of Balls Faced vs Runs per Ball')

# Plot 3: Combined plot with average line
plt.figure(figsize=(10, 6))

# Plotting the original data
plt.scatter(cric_df['Number of Balls Faced'], cric_df['Runs Per Ball'], label='Original Data points', alpha=0.5)
#plt.scatter(cric_df['Random Balls Faced'], cric_df['Random Runs per Ball'], label='Random Data points', alpha=0.5)

# Smooth curve for original data
smoothed_orig = lowess(cric_df['Runs Per Ball'], cric_df['Number of Balls Faced'], frac=0.3)
plt.plot(smoothed_orig[:, 0], smoothed_orig[:, 1], color='blue', label='Original Smooth curve')

# Smooth curve for random data
smoothed_rand = lowess(cric_df['Random Runs per Ball'], cric_df['Random Balls Faced'], frac=0.3)
plt.plot(smoothed_rand[:, 0], smoothed_rand[:, 1], color='red', label='Random Smooth curve')

# Calculate and plot average line for original data
avg_runs_per_ball = cric_df['Runs Per Ball'].mean()
plt.axhline(y=avg_runs_per_ball, color='green', linestyle='--', label='Average Runs per Ball (Original)')

plt.title('Combined Plot: Balls Faced vs Runs per Ball')
plt.xlabel('Balls Faced')
plt.ylabel('Runs Per Ball')
plt.legend()
plt.grid(True)
# plt.savefig('Combined Plot: Balls Faced vs Runs per Ball.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.nonparametric.smoothers_lowess import lowess
import scipy.stats as stats

# Function to plot data with a smooth curve using LOWESS
def plot_with_curve(x, y, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, label='Data points')
    
    # Apply LOWESS smoothing
    smoothed = lowess(y, x, frac=0.3)
    
    plt.plot(smoothed[:, 0], smoothed[:, 1], color='red', label='Smooth curve')
    plt.title(title)
    plt.xlabel('Balls Faced')
    plt.ylabel('Runs per Ball')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return smoothed

# Function to plot residuals
def plot_residuals(residuals, title):
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    plt.hist(residuals, bins=30, edgecolor='k')
    plt.title(f'Histogram of Residuals: {title}')
    plt.xlabel('Residuals')
    plt.ylabel('Frequency')
    
    plt.subplot(1, 2, 2)
    stats.probplot(residuals, dist="norm", plot=plt)
    plt.title(f'Q-Q Plot of Residuals: {title}')
    
    plt.tight_layout()
    plt.show()

# Function to process and plot data for a given gender
def process_and_plot_by_gender(df, gender):
    df_gender = df[df['Gender'] == gender]
    
    # Plot 1: Random Balls Faced vs Random Runs per Ball
    smoothed_rand = plot_with_curve(df_gender['Random Balls Faced'], df_gender['Random Runs per Ball'], 
                                    f'{gender} - Random Balls Faced vs Random Runs per Ball')
    
    # Plot 2: Number of Balls Faced vs Runs per Ball
    smoothed_orig = plot_with_curve(df_gender['Number of Balls Faced'], df_gender['Runs Per Ball'], 
                                    f'{gender} - Number of Balls Faced vs Runs per Ball')
    
    # Calculate residuals
    residuals_rand = df_gender['Runs Per Ball'] - np.interp(df_gender['Random Balls Faced'], 
                                                            smoothed_rand[:, 0], smoothed_rand[:, 1])
    residuals_orig = df_gender['Runs Per Ball'] - np.interp(df_gender['Number of Balls Faced'], 
                                                            smoothed_orig[:, 0], smoothed_orig[:, 1])
    
    # Plot residuals for random data
    plot_residuals(residuals_rand, f'{gender} - Random Data')
    
    # Plot residuals for original data
    plot_residuals(residuals_orig, f'{gender} - Original Data')

# Process and plot for male data
process_and_plot_by_gender(cric_df, 'male')

# Process and plot for female data
process_and_plot_by_gender(cric_df, 'female')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import shapiro, anderson

# Load your data
df = cric_df

# Statistical tests for normality
def test_normality(data, column_name):
    # Shapiro-Wilk Test
    stat, p = shapiro(data)
    print(f'Shapiro-Wilk Test for {column_name}: Statistics={stat:.3f}, p={p:.3f}')
    if p > 0.05:
        print(f'{column_name} looks Gaussian (fail to reject H0)')
    else:
        print(f'{column_name} does not look Gaussian (reject H0)')
    
    # Anderson-Darling Test
    result = anderson(data)
    print(f'Anderson-Darling Test for {column_name}: Statistic: {result.statistic:.3f}')
    for i in range(len(result.critical_values)):
        sl, cv = result.significance_level[i], result.critical_values[i]
        if result.statistic < cv:
            print(f'{column_name}: {sl:.3f}%: {cv:.3f}, data looks normal (fail to reject H0)')
        else:
            print(f'{column_name}: {sl:.3f}%: {cv:.3f}, data does not look normal (reject H0)')

# Test for Column1
test_normality(df['Number of Balls Faced'], 'Number of Runs Scored')

# Test for Column2
test_normality(df['Runs Per Ball'], 'Runs Per Ball' )

# Q-Q plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
stats.probplot(df['Number of Balls Faced'], dist="norm", plot=plt)
plt.title('Q-Q plot for Number of Runs Scored')

plt.subplot(1, 2, 2)
stats.probplot(df['Runs Per Ball'], dist="norm", plot=plt)
plt.title('Q-Q plot for Runs Per Ball')
plt.tight_layout()
plt.show()

# Histograms
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['Number of Runs Scored'], bins=30, edgecolor='k')
plt.title('Histogram for Number of Runs Scored')

plt.subplot(1, 2, 2)
plt.hist(df['Runs Per Ball'], bins=30, edgecolor='k')
plt.title('Histogram for Runs Per Ball')
plt.tight_layout()
plt.show()
```
