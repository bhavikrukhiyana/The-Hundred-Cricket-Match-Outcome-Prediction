```python
# Create a dictionary from the lists
dict_data = {
    'Date': date,
    'Match Number': match_num,
    'Season': season,
    'Gender':gender,
    'Innings': innings,
    'Toss Winner': toss_winner,
    'Toss Decision': toss_dec,
    'Match Winner': match_winner,
    'Venue':venue,
    'Player Team': player_team,
    'Team Against': against_team,
    'target': target,
    #'Team Score': team_scores,
    'Player Name':player_name,
    'Number of Runs Scored': nrs,
    'Number of Balls Faced': nbf,
    '4s': bt_n4s,
    '6s': bt_n6s,
    'Strike Rate': str_rate,
    'Wicket Type': wtype,
    'Number of Balls Bowled': nbb,
    'Number of Runs Conceded': nrgs,
    'Economy': economy,
    'Catches': ct,
    'Run Outs': rot,
    'Wicket Taken': wt
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(dict_data)
df['Date'] = pd.to_datetime(df['Date'])
# Display the DataFrame
df.dtypes
```

```python
m_id = []
inning = []
gen = []
sea = []
team_scores = []
for filename in os.listdir(os.getcwd()):
    if filename.endswith('.json'):
        #print(filename)
        with open(filename,'r') as file:
            json_data = file.read()
            data = json.loads(json_data)
            if 'match_number' not in data["info"]['event']:
                        #print(data["info"]['event']['stage'])
                match_id = data["info"]['event']['stage']
            else:
                        #print(data["info"]['event']['match_number'])
                match_id= data["info"]['event']['match_number']
            if len(data['innings']) > 1:
                for index, inn in enumerate(data['innings']):
                    team_score = 0
                    for i in inn['overs']:
                        for j in i['deliveries']:
                            team_score += j['runs']['total']
                    inning.append(index+1)
                    gen.append(data["info"]['gender'])
                    sea.append(data["info"]['season'])
                    team_scores.append(team_score)
                    m_id.append(match_id)

```

```python
print(len(m_id))
print(len(gen))
print(len(sea))
print(len(team_scores))
print(len(inning))


match_stats = {
    "match_id": m_id,  
    "gender": gen,
    "season": sea,
    "team_scores": team_scores,
    "inning": inning
}

# Create the DataFrame using pd.DataFrame()
match_stats = pd.DataFrame(match_stats)

# Print the DataFrame
print(match_stats.info())
```
